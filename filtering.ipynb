```python
import pandas as pd
```

```python
DATASET1 ='data/processed/train.csv'
DATASET2 ='data/processed/test.csv'

OUTPUT1 = DATASET1.replace('.csv', '_clean.csv')
OUTPUT2 = DATASET2.replace('.csv', '_clean.csv')
```

```python
df1 = pd.read_csv(DATASET1)
df2 = pd.read_csv(DATASET2)
```

```python
to_save_cols = []

# Verificar colunas de mesmo nome nos dois
for col in df1.columns:
    if col in df2.columns:
        to_save_cols.append(col)
```

```python
# Filtrar dois datasets com as colunas em comum
df1 = pd.concat([df1[to_save_cols], df1['SalePrice']], axis=1)
df2 = df2[to_save_cols]
```

```python
# Salvar datasets
df1.to_csv(OUTPUT1, index=False)
df2.to_csv(OUTPUT2, index=False)
```
